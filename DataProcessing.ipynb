This purpose of this project is to use collaborative filtering techniques to predict user scores of top rated anime, from a dataset I acquired from myanimelist.net a few years ago via the BeautifulSoup Python package.

myanimelist.net has since changed some of the structure of its site HTML, so the code to acquire the data doesn't work anymore, but I have included the code in a file named datacollection.py.

The data relating to approximately 2000 of the top anime series at the time (including movies) is contained in animes3.csv.

The scoring data is contained in a folder called newUsersV2, with the scores for each user contained in a separate file.

First, I will import the necessary packages. I will be splitting the data into training and test sets, and doing a lot of processing in pandas.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

I load the anime data and print the head.

```python
animes = pd.read_csv('animes3.csv')
print(animes.head())
```

I provide each anime with a unique ID for easier referencing.

```python
animes['animeID'] = np.arange(len(animes))

print(animes.head())
```

Next, I collect the user data, and combine it with the animes dataframe to get unique anime, user, score entries.

I combine all this data in a dataframe called allusers.

```python
animes = animes.iloc[:,1:]
```

```python
animes.to_csv('animes.csv')
```

```python
animes = pd.read_csv('animes.csv')
allusers = pd.DataFrame(columns=['Name','Score','userID','animeID'])

for i in range(2155):
    try:
        df = pd.read_csv('newUsersV2/User_' + str(i) + '.csv')
        df = df[['Name','Score']]
        df['userID'] = i
        df = df[df['Score'] <= 10]
        df = df[df['Score'] >= 1]
        df = df.merge(animes[['Name','animeID']],how='inner',on='Name')
        #print(df.head())
        allusers = allusers.append(df)
    except:
        print("bla")

print(allusers.shape)
```

Here, I convert the anime categorical variables to dummy variables using one-hot encoding, and store the new data in a dataframe called animes_dummies.

```python
animes['Genres'] = animes['Genres'].str.split('/')
print(animes['Genres'])
allgenres = set(animes['Genres'].sum())
print(allgenres)

for genre in allgenres:
    animes[genre] = animes['Genres'].apply(lambda x: 1 if genre in x else 0)
    
print(animes.columns)
animes = animes.drop(['Genres'],axis=1)
animes_dummies = animes.iloc[:,1:]
animes_dummies = pd.get_dummies(animes_dummies)

animes_dummies['Very Popular'] = animes_dummies['Popularity'].apply(lambda x: 1 if x >= 250000 else 0)
animes_dummies['Popular'] = animes_dummies['Popularity'].apply(lambda x: 1 if (x < 250000 and x >= 50000) else 0)
animes_dummies['Obscure'] = animes_dummies['Popularity'].apply(lambda x: 1 if (x < 50000 and x >= 10000) else 0)
animes_dummies['Very Obscure'] = animes_dummies['Popularity'].apply(lambda x: 1 if x < 10000 else 0)

animes_dummies = animes_dummies.drop(['Popularity'],axis=1)
print(animes_dummies.head())
```

```python
animes_dummies.to_csv('animes_dummies.csv')
```

Here, I separate the user-anime-rating data into a train set and a test set.

```python
allusers = pd.read_csv('allusers.csv',index_col=0)
print(allusers.head())

X = allusers[['userID','animeID']]
y = allusers['Score']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=24)
```

I temporarily combine X_train and y_train in order to build my model.

```python
train_data = X_train
train_data['Score'] = y_train
```

In the next few steps, I construct a dataframe called user_means_stds, which contains the mean and standard deviation for each user's scores, to make standardization easier later.

I also convert the train data to a dataframe consisting of standardized attribute values. These attributes consist of those in animes_dummies, and the means corresponding to each attribute for each user. Each user-anime pair will be replaced with these combined attributes, which will be used to evaluate the content-based filter, one of the models I will be using to predict the test values.

```python
users = train_data['userID'].unique()
count = 0
for user in users:
    userdf = train_data[train_data['userID'] == user]
    #print(userdf)
    userdf = userdf.merge(animes_dummies,how='inner',on='animeID')
    #print(userdf.head())
    for column in userdf.columns[4:]:
        #print(userdf['Score'])
        colnew = column + '_avg'
        userdf[colnew] = userdf['Score'][userdf[column] == 1].mean()
        #print(userdf[colnew])
        userdf[colnew] = userdf[colnew].fillna(userdf['Score'].mean())
    userdf.to_csv('userdfsnew/userdf_' + str(count) + '.csv')
    count +=1
    print(userdf[colnew].head())
    print(count)
```
